```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/AI GO/深度學習/data/training_data.csv')
data
```

```python
import statsmodels.api as sm
import pandas as pd
```

```python
# 拟合线性回归模型
model = sm.OLS(data['單價'], data[['土地面積', '移轉層次', '總樓層數', '屋齡', '建物面積', '車位面積', '車位個數', '橫坐標', '縱坐標', '主建物面積', '陽台面積', '附屬建物面積']])  # 替换自变量名称
results = model.fit()

# 查看模型摘要信息
print(results.summary())
```

```python
# 添加一个截距项（intercept）
data['intercept'] = 1

# 拟合线性回归模型
model = sm.OLS(data['單價'], data[['intercept', '土地面積', '移轉層次', '總樓層數', '屋齡', '建物面積', '車位面積', '車位個數', '橫坐標', '縱坐標', '主建物面積', '陽台面積', '附屬建物面積']])  # 替换自变量名称
results = model.fit()

# 查看模型摘要信息
print(results.summary())
```
